
Linear regression using two methods
===
In this notebook, I illustrate two different methods to perform linear regression (i.e. supervised learning) on a simple dataset. I build a linear regression model to predict profits for a food truck based on population size of a city. I intentionally omit to split the data into a training-validation-test sets for this project. More complete analyses will be included in different projects. Although this dataset is small and has only one feature, the two methods I illustrate here can be used with larger datasets including more than one feature. I describe the two methods I use in the following two parts: 

1. Linear regression using the traditional ordinary least squared (OLS) method
2. Linear regression using Machine Learning (detailed version with calculations)

Part 1 - Linear regression using the traditional ordinary least squared (OLS) method
---

```python
import scikits.statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data
house = np.loadtxt('./Data/ex1data1.txt', delimiter=',')
print house[:10]
print '\n dimension:', house.shape
print '\n', type(house)
```

```python
# prepare the data
X = house[:, :1] # X.shape = (97, 1)
y = house[:, -1] # y.shape = (97, 1)

print '\n X = \n', X[:5]
print '\n y = \n', y[:5]
```

```python
# run the OLS model
lm_model = sm.OLS(y,sm.add_constant(X)).fit()

print lm_model.summary()
```

> A few things to note: Results indicate a positive relationship, with food truck profits increasing (by 1.19) with population. This linear model is statistically significant as indicated by the small p-value (p${<}$0.05) and large F-statistic. The model explains about 70% of the variance as indicated by the R-squared value. The Durbin-Watson value is high (0.99) and suggests that the errors are independent (i.e. no autocorrelation). The data tend to be spread out more to the right (i.e. positive Skew) and the high Kurtosis (${>}$3) reflects a leptokurtic (i.e. fat) distribution.

> A quick note about plotting: they can be done in two ways: 

- Implicit ways are the simplest and fastest way to plot figures. They usually uses plot() and are great for interactive plotting such as plotting in IPython notebooks
- Explicit ways are more detailed and lists out all the steps to set up the graphs. They are preferable for non-interactive plotting such as scripting

>In this notebook, I choose to code my plots using the explicit method.

```python
# visualize the data
fig = plt.figure(figsize=(15, 5))

## left plot: data with regression line
ax = fig.add_subplot(121) # 1 row, 2 columns, and show the first subplot

ax.scatter(X, y, marker = 'x', color='red')

a1 = np.arange(5, 25)
b1 = lm_model.params[1] + lm_model.params[0] * a1
ax.plot(a1, b1)

ax.set_xlabel('Population of city in 10,000s')
ax.set_ylabel('Profit in $10,000s')

ax.set_xlim(5, 25)
ax.set_ylim(-5, 25)

ax.set_title("Scatter plot of training data")

## right plot: residual plot
ax = fig.add_subplot(122) # 1 row, 2 columns, and show the second subplot

ax.scatter(lm_model.fittedvalues, lm_model.resid, color = 'blue')
ax.hlines(0, 0, 25, linestyles = 'dashed', color = 'red')

ax.set_xlabel('Fitted values')
ax.set_ylabel('Residuals')

ax.set_xlim(0, 25)

ax.set_title("Residual plot")

plt.show()
```

> Note: The residual plot somehow reinforce the observations made in the regression output. There seem to be a pattern in the residual plot and one outlier can be spotted to the upper left. Transforming the data (e.g. log transform) prior to re-running the model might improve the results.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data
house = np.loadtxt('./Data/ex1data1.txt', delimiter=',')
print house[:10]
print '\n dimension:', house.shape
print '\n', type(house)
```

```python
# visualize the data
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(house[:,0], house[:, 1], marker = 'x', color='red')

ax.set_xlabel('Population of city in 10,000s')
ax.set_ylabel('Profit in $10,000s')

ax.set_xlim(5, 25)
ax.set_ylim(-5, 25)

ax.set_title("Scatter plot of training data")
plt.show()
```

## Some background:

As a reminder, the **linear regression hypothesis** is defined as:

$$
h_{\theta}(x) = {\theta}^Tx = {\theta}_0 + {\theta}_1x_1 + ... + {\theta}_nx_n
$$

The objective of the linear regression is to estimate the parameters of the model (${\theta}_j$ values) by minimizing the following **cost function**:

$$
J{(\theta)} = \frac{1}{2m} \sum_{i=1}^{m}(h_{\theta}(x^{(i)}) - y^{(i)})^2
$$

The **batch gradient descent algorithm** can be used to minimize the cost function $J{(\theta)}$ and is given by the following: 

$$
{\theta}_j - {\alpha} \frac{\partial J(\theta)}{\partial \theta_{j}}
$$
where
$$
\frac{\partial J(\theta)}{\partial \theta_{j}} = \frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x^{(i)}) - y^{(i)}) x^{(i)}_{j}
$$

```python
# Prepare the data
add_ones = np.ones(shape=(len(house), 3)) # add_ones.shape = (97,3)
add_ones[:, 1:] = house

X = add_ones[:, :2] # X.shape = (97, 2)
y = add_ones[:, 2] # y.shape = (97, )

m = len(y)

print 'add_ones = \n', add_ones[:5]
print '\n X = \n', X[:5]
print '\n y = \n', y[:5]
```

```python
# initialize the thetas at zero, set the number of iterations and the learning rate alpha
theta = np.zeros(shape=(2, 1)) # theta.shape = (2, 1)
iterations = 1500
alpha = 0.01
```

```python
# define the cost function
def cost_function(X, y, theta):
    '''
    compute the cost function J and calculate the gradient at initial theta(zeros)
    '''
    h = X.dot(theta).flatten() # hypothesis; h.shape = (97,)
    
    J = (1./(2.*m)) * sum((h - y)**2) # cost function
    
    #grad = (1./m) * X.T.dot(h - y) # gradient at initial theta(zeros)
    
    return J
```

```python
# calculate the cost function at initial theta(zeros)
cost_function(X, y, theta)
```

```python
# define the gradient descent algorithm
def grad_descent(X, y, theta, alpha, iterations):
    '''
    computed the batch gradient descent algorithm given the number of iterations and the learning rate alpha
    '''
    J_history = np.zeros(shape=(iterations, 1))
    
    for i in range(iterations):
        h = X.dot(theta).flatten() # hypothesis; h.shape = (97,)    
        grad = (1./m) * X.T.dot(h - y) # gradient descent
        theta = theta.flatten() - alpha * grad
        
        J_history[i, 0] = cost_function(X, y, theta)
        
    return theta, J_history
```

```python
# gradient descent update
grad_descent(X, y, theta, alpha, iterations)
```

> According to the above results, the cost function returned by the gradient descent algorithm after 1500 iterations (alpha = 0.01) is: $J({\theta}) = 4.48$

> and the linear regression equation is: $y = -3.63 + 1.17x_1$

```python
# Add the linear regression line to the scatter plot
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(house[:,0], house[:, 1], marker = 'x', color='red', label = 'Training data')

a2 = np.arange(5, 25)
b2 = -3.63 + 1.17 * a2
ax.plot(a2, b2, label = 'Linear regression')

ax.set_xlabel('Population of city in 10,000s')
ax.set_ylabel('Profit in $10,000s')

ax.set_xlim(5, 25)
ax.set_ylim(-5, 25)

ax.set_title("Scatter plot of training data")
ax.legend(loc = 'lower right')
plt.show()
```

```python
# Plotting the cost function as a function of the number of iterations
fig = plt.figure()
ax = fig.add_subplot(111)

a3 = np.arange(iterations)
cf = grad_descent(X, y, theta, alpha, iterations)
b3 = cf[1]
ax.plot(a3, b3)

ax.set_xlabel('Number of iterations')
ax.set_ylabel('Cost function J(theta)')

ax.set_xlim(-100, 1600)
ax.set_ylim(4.4, 7)

ax.set_title("Gradient descent algorithm for linear regression")
plt.show()
```

> Note: 1500 iterations with a learning rate alph = 0.01 seems to be appropriate. The cost function converges to the local minimum. The parameters thetas are very similar to the ones found by the OLS model in Part 1.
