Web client login behavior
===
This project uses web client login time for the month of March 2012. The goal of this project is to explore the data set and generate graphs showing the breakdown of client login behavior. I explored the same data set in R. You can find it under my "Data_Analysis_R" Github repo ([User Behavior.html](https://github.com/Prim8/Data_Analysis_R))

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_rows', 200)
pd.version.version
#sys.version
```

```python
# Read json file
logs = pd.read_json('logs.json', convert_dates=[0], typ='Series')
logs
```

```python
#logs.to_csv('logs.csv') # save it to csv for further analyses in R
```

```python
# convert to time series
dti = pd.DatetimeIndex(logs)
dti
```

```python
# get the daily breakdown of logs
day = logs.groupby(dti.day).size()
day
```

```python
# plot the daily breakdown of logs
day.plot()
xlabel('Day of the month of March')
ylabel('Number of logins')
title('Daily breakdown of logins')
```

```python
# get the hourly breakdown of logs
hour = logs.groupby(dti.hour).size()
hour
```

```python
# plot the hourly breakdown of logs
hour.plot()
xlabel('Hour of the day')
ylabel('Number of logins')
title('Hourly breakdown of logins')
```

```python
# get the day of the week breakdown of logs (Monday=0, Sunday=6)
day_of_week = logs.groupby(dti.dayofweek).size()
day_of_week
```

```python
# plot the day of the week breakdown of logs (Monday=0, Sunday=6)
day_of_week.plot()
xlabel('Day of the week')
ylabel('Number of logins')
title('Day of the week breakdown of logins')
```

```python
# side by side plots
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(day)
axes[0].set_xlabel('Day of the month of March')
axes[0].set_ylabel('Number of logins')
axes[0].set_title('Daily breakdown of logins')

axes[1].plot(hour)
axes[1].set_xlabel('Hour of the day')
axes[1].set_title('Hourly breakdown of logins')

axes[2].plot(day_of_week)
axes[2].set_xlabel('Day of the week')
axes[2].set_title('Day of the week breakdown of logins')
```
