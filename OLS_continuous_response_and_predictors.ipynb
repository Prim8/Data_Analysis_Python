# OLS with continuous response and predictors

<br>

> **Modules**

```python
%matplotlib inline

import pandas as pd
import statsmodels
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats
```

<br>

> **Data**

Outcome:    
* third column: price of the house

Predictors: 
* first column: size of the house (in square feet)
* second column: number of bedrooms

```python
# load data (housing price)
col_names = ['size', 'beds', 'price']
df = pd.read_table('./Data/ex1data2.txt', sep=',', names=col_names)
print df.head()
```

```python
# assign y and X
y = df.price
X = df.ix[:, :2]
print y.head()
print
print X.head()
```

<br>

> **Fit OLS model**

```python
# run the OLS model
model = sm.OLS(y, sm.add_constant(X)).fit()
print model.summary()
```

```python
print 'R2: ', model.rsquared
print
print 'Parameters: ', model.params
print
print 'Standard errors: ', model.bse
print
#print 'Predicted values: ', model.predict()
```

<br>

> **Diagnostics**

Rainbow test for linearity:

* $H_0$: the relationship is linear
* $H_A$: the relationship is not linear

small p-value => the relationship is not linear (reject $H_0$)

```python
rainbow = sm.stats.linear_rainbow(model)
print 'F-statistic = %6.4f, p-value = %6.4f' % rainbow
print
if rainbow[1] <= 0.05:
    print 'Decision:\n reject H0: the relationship is not linear.'
else:
    print 'Decision:\n accept H0: the relationship is linear.'
```

<br>

> **Plots**

```python
fig = plt.figure(figsize=(10, 4))

## upper left plot: Residuals vs Fitted
ax = fig.add_subplot(121) # 1 row, 2 columns, and show the first subplot

ax.scatter(model.fittedvalues, model.resid)

ax.plot([min(model.fittedvalues), max(model.fittedvalues)], [0.0, 0.0], 'r-')

ax.set_title('Residuals vs Fitted')
ax.set_ylabel('Residuals')
ax.set_xlabel('Fitted values')
plt.grid()

## upper right plot: Normal Q-Q
ax = fig.add_subplot(122) #  row, 2 columns, and show the second subplot

statsmodels.graphics.gofplots.qqplot(model.resid, fit=True, line='45', ax=ax)

ax.set_title('Normal Q-Q')
ax.set_ylabel('Standardized Residuals')
#ax.set_xlabel('Theoretical Quantiles')
plt.grid()

plt.tight_layout()
```

<br>

> **Predictions**

```python
model.predict([1, 1600, 4])
```
